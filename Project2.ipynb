# Question2 - preprocessing & feature engineering

```python
missing_data = dataframe.isna()   #detect missing values
missing_data_count = missing_data.loc[:,:].sum()

missing_data_count
```

```python
#fill missing data with average 
Arrival_Delay_in_Minutes_mean = dataframe['Arrival Delay in Minutes'].mean() 
dataframe['Arrival Delay in Minutes'].fillna(Arrival_Delay_in_Minutes_mean, inplace = True)
missing_data = dataframe.isna()   #detect missing values
missing_data_count = missing_data.loc[:,:].sum()

missing_data_count
```

```python
#Detect outliers data
Q1 = dataframe.quantile(0.25)
Q3 = dataframe.quantile(0.75)
IQR = Q3 - Q1
# Removing outliers 
dataframe = dataframe[~((dataframe < (Q1 - 1.5 * IQR)) |(dataframe > (Q3 + 1.5 * IQR))).any(axis=1)]
dataframe
```

```python
#convert categorical value to numerical value.
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
dataframe['Gender'] = labelencoder.fit_transform(dataframe['Gender'])
dataframe['Customer Type'] = labelencoder.fit_transform(dataframe['Customer Type'])
dataframe['Type of Travel'] = labelencoder.fit_transform(dataframe['Type of Travel'])
dataframe['Class'] = labelencoder.fit_transform(dataframe['Class'])
dataframe['satisfaction'] = labelencoder.fit_transform(dataframe['satisfaction'])
dataframe
```

# Question3

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

input_data = dataframe.iloc[:,:-1]
output_target=dataframe.iloc[:, 22]

X_train, X_test, y_train, y_test = train_test_split(input_data, output_target, test_size = 0.3, random_state=42)



model = DecisionTreeClassifier(criterion='gini')

model.fit(X_train, y_train)

y_prediction = model.predict(X_test)



result = confusion_matrix(y_test,y_prediction)

print("Confusion : ",result)
report = classification_report(y_test, y_prediction)
print("Result:",report)


```
